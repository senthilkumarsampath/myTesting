```python
import docx
import re

def extract_author_date(doc_path, output_path):
    doc = docx.Document(doc_path)
    text = '\n'.join([para.text for para in doc.paragraphs])
    author_date_pattern = "(?:[A-Z][a-z]+?\s(?:and|&))?\s?(?:[A-Z][a-z]+)\s?(?:et\.?\sal\.?)?\s?(?:\,)?\s\d+"
    # author_date_pattern = r"\(([A-Z][a-z]+(?:(?: *[A-Z][a-z]+)*)?)(?: *(?:et\.? *al\.)|(?:(?:, *(?:and|&) *)([A-Z][a-z]+(?: *[A-Z][a-z]+)*)))?),? *([0-9]{4})\)"

    author_date_match = re.findall(author_date_pattern, text)

    for author in author_date_match:
        print(author)

    with open(output_path, "w") as f:
        for match in author_date_match:
            f.write(" ".join(filter(None, match)) + '\n')

extract_author_date("/Users/senthil/Desktop/Senthil/myTesting/python_scripts/word_automation/RL_08_WOLI_C001.docx", "author_date_citations.txt")
```
