```python
import re
import docx
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the Word document
doc = docx.Document("/Users/senthil/Desktop/Senthil/myTesting/Project_Creation/PM Brief Form_From Action to Ethics.docx")

# Extract text from paragraphs in the document
extracted_text = ""

extracted_info = {
    'book_title': 'Frictionlessness',
    'subtitle': 'The Silicon Valley Philosophy of Seamless Technology and the Aesthetic Value of Imperfection',
    'series': 'Thinking Media',
    'author': 'Dr Jakko Kemper',
    'author_mailid': 'J.Kemper@uva.nl',
    'isbn': ':',
    'hardback_isbn': '9798765104415',
    'paperback_isbn': '9798765104422',
    'epub_isbn': '9798765104453',
    'epub2_isbn': '9798765104439',
    'ebook_master': '9798765104446',
    'project_endDate': '15 September 2023',
    'production_editor': 'Amy Brownbridge',
    'text_design': 'Monograph C',
    'trim_size': '5.5 x 8.5',
    'Indexer': 'Project Manager'
}

for paragraph in doc.paragraphs:
    extracted_text += paragraph.text + "\n"

# Preprocess the text
extracted_text = re.sub(r'\b\w{1,2}\b', '', extracted_text)  # Remove words with two or fewer characters

# Define the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit the vectorizer and transform the extracted text
tfidf_matrix = vectorizer.fit_transform([extracted_text])

# Get the feature names (terms)
feature_names = vectorizer.get_feature_names_out()

# Extract the top TF-IDF terms and their corresponding weights
top_terms = []
top_weights = []
top_k = 10  # Number of top terms to extract

# Iterate over the features and their weights
for feature, weight in zip(feature_names, tfidf_matrix.toarray()[0]):
    top_terms.append(feature)
    top_weights.append(weight)

# Get the top K terms with highest weights
top_terms = [term for _, term in sorted(zip(top_weights, top_terms), reverse=True)][:top_k]

# Extract the relevant information using the top terms
extracted_information = {}
for key in extracted_info:
    extracted_information[key] = []  # Initialize list for each key

    for term in top_terms:
        if term.lower() in extracted_info[key].lower() and extracted_info[key] not in extracted_information[key]:
            extracted_information[key].append(extracted_info[key])

# Print the extracted information
for key, values in extracted_information.items():
    print(f"{key}: {values}")
```

```python
from docx import Document

def extract_paragraphs_with_numbering(document_path):
    document = Document(document_path)
    
    list_paragraphs = []
    for paragraph in document.paragraphs:
        if paragraph.style.name.startswith('List') and paragraph.text.strip():
            list_paragraphs.append(paragraph.text.strip())
    
    return list_paragraphs

# Usage example
document_path = '/Users/senthil/Desktop/Senthil/myTesting/Project_Creation/RL_01_CLED_C001.docx'
result = extract_paragraphs_with_numbering(document_path)
for i, item in enumerate(result, start=1):
    print(f"Label {i}: {item}")
```
