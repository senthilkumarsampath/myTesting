```python
import xml.etree.ElementTree as ET
import csv

# parse the XML file
tree = ET.parse('/Users/senthil/Desktop/Senthil/myTesting/Project_Creation/9781032119557_prepress_brief_1669900724.xml')
root = tree.getroot()

# create a CSV file and write the headers
with open('file.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['label', 'attribute', 'value'])
    
    # iterate through the root elements and their children
    for elem in root.iter():
        # write the tag name as the label
        label = elem.tag
        # write the attributes and values as attribute and value
        for attr, value in elem.attrib.items():
            writer.writerow([label, attr, value])
        # write the text content as value
        writer.writerow([label, '', elem.text])
```

```python
from lxml import etree
import pandas as pd

# Parse XML file
xml_file = '/Users/senthil/Desktop/Senthil/myTesting/Project_Creation/9781032119557_prepress_brief_1669900724.xml'
tree = etree.parse(xml_file)
root = tree.getroot()

# Create a list to store the data
data = []

# Iterate through all elements in the XML
for element in root.iter():
    # Extract tag, attributes, and text
    tag = element.tag
    attributes = element.attrib
    text = element.text
    
    # Append data to list
    data.append([tag, attributes, text])

# Create a DataFrame from the list
df = pd.DataFrame(data, columns=['Tag', 'Attributes', 'Text'])
# df.to_csv('data.csv', index=False)
# Print the DataFrame
# pd.options.display.max_columns = None
# pd.options.display.max_rows = None
print(df)
```
